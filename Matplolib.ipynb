```python
#Matplolib
import matplotlib.pyplot as plt
```

```python
dias = ['Seg', 'Ter', 'Quar', 'Qui', 'Sex', 'Sab', 'Dom']
temperaturas = [19,22,23,21,27,31,33]
```

```python
plt.figure(figsize=(6,4))
plt.plot(dias, temperaturas, marker='x', linestyle='--', color='red')
plt.title("Evolução da Temperatura ao longo da semana")
plt.xlabel("Dias da semana")
plt.ylabel("Temperatura (°C)")
plt.grid(True, linestyle=":", alpha=0.2)
plt.show()
```

```python
meses = ['Abr', 'Mar', 'Jun', 'Jul', 'Ago', 'Set']
Color = [5,14,25,20,19,22]
Itamar = [10,15,21,16,14,19]
FHC = [20,22,25,28,32,36]
plt.figure(figsize=(10,6))
plt.plot(meses, Color, marker='o', linestyle='--', color='green', label='Fernando Collor')
plt.plot(meses, Itamar, marker='x', linestyle='-', color='blue', label='Itamar Franco')
plt.plot(meses, FHC, marker='^', linestyle='-.', color='red', label='FHC')
plt.title("Intenções de voto para presidente")
plt.xlabel("Mês da Pesquisa")
plt.ylabel("Intenção de Votos (%)")
plt.ylim(0,50)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Bebidas']
vendas = [120,95,210,80]
```

```python
plt.figure(figsize=(8,4))
plt.bar(categorias, vendas, color=['red','green','orange','blue'])
plt.title("Vendas por categoria de produtos")
plt.xlabel("Categoria de Produtos")
plt.ylabel("Quantidade vendida")
plt.ylim(0,240)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import numpy as np
np.random.seed(42)
notas = np.random.normal(70,10,500)
```

```python
plt.figure(figsize=(8,5))
plt.hist(notas,bins=50, edgecolor='black', color='green')
plt.title("Distribuição de notas da turma")
plt.xlabel("Notas")
plt.ylabel("Número de alunos")
plt.axvline(notas.mean(),color='red', linestyle='--', linewidth='2')
plt.legend(['Média das Notas'])
plt.grid(True, linestyle=':', alpha=0.5)
plt.ylim(0,35)
plt.show()
```

```python
ativos = ['Ações', 'Renda Fixa', 'Ouro', 'Criptomoedas']
valores = [45,30,15,10]
plt.figure(figsize=(7,7))
explode = (0.1,0,0,0)
plt.pie(valores, labels=ativos, startangle=90, autopct='%1.1f%%', explode=explode)
plt.title("Composição da carteira de investimentos")
plt.axis('equal')
plt.show()
```

```python
np.random.seed(0)
alturas = np.random.normal(170,10,50)
pesos = (alturas * 0.5) + np.random.normal(0,5,50)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(alturas, pesos, color='purple', s=20)
plt.title("Relação entre altura e peso")
plt.xlabel("Altura (cm)")
plt.ylabel("Peso(kg)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
